# Figure 1c

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

alpha_vals = np.linspace(0, 1, 200)      # x-axis: alpha in [0, 1]
c_vals     = np.linspace(0, 10, 200)     # y-axis: c  (linear, adjust range as needed)

# Single-class "generalizes" field
field = np.ones((len(c_vals), len(alpha_vals)))

fig, ax = plt.subplots(figsize=(7, 5))

cmap = ListedColormap(['#7FBF7F'])        # green
ax.pcolormesh(
    alpha_vals, c_vals, field,
    cmap=cmap, shading='auto'
)

# Red line at alpha = 1 (generalization fails)
ax.axvline(1.0, color='crimson', lw=3.0, zorder=3)

ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$c$')
ax.set_title('Transitive inference, kernel model:\nphase diagram of generalization')
ax.set_xlim(0, 1.02)
ax.set_ylim(c_vals[0], c_vals[-1])

legend_elems = [
    Patch(facecolor='#7FBF7F', edgecolor='none',
          label=r'Generalizes  ($\alpha < 1$)'),
    Line2D([0], [0], color='crimson', lw=3,
           label=r'Fails  ($\alpha = 1$)')
]
ax.legend(handles=legend_elems, loc='upper left', framealpha=0.95)

plt.tight_layout()
plt.show()
fig.savefig("figure_2c.pdf", bbox_inches="tight")
fig.savefig("figure_2c.svg", bbox_inches="tight")
```

# Figure 1d

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

alpha_vals = np.linspace(0, 1, 200)      # x-axis: alpha in [0, 1]
c_vals     = np.linspace(0, 10, 200)     # y-axis: c (linear)

# Single-class "memorizes" field
field = np.ones((len(c_vals), len(alpha_vals)))

fig, ax = plt.subplots(figsize=(7, 5))

cmap = ListedColormap(['#7FBF7F'])        # green
ax.pcolormesh(
    alpha_vals, c_vals, field,
    cmap=cmap, shading='auto'
)

# Red line at alpha = 0 (memorization fails)
ax.axvline(0.0, color='crimson', lw=3.0, zorder=3)

ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$c$')
ax.set_title('Transverse patterning, kernel model:\nphase diagram of memorization')
ax.set_xlim(-0.02, 1.0)
ax.set_ylim(c_vals[0], c_vals[-1])

legend_elems = [
    Patch(facecolor='#7FBF7F', edgecolor='none',
          label=r'Memorizes  ($\alpha > 0$)'),
     
    Line2D([0], [0], color='crimson', lw=3,
           label=r'Fails  ($\alpha = 0$)')
]
ax.legend(handles=legend_elems, loc='upper right', framealpha=0.95)

plt.tight_layout()
plt.show()
fig.savefig("figure_2d.pdf", bbox_inches="tight")
```
