```python
from Ranking_exp_test import Ranking_exp
import numpy as np
import matplotlib.pyplot as plt

#plot ranking systems, vary exception position, while keeping n constant

n_val = list(range(7, 16, 1))
excep_len = 3
k_o_vals = np.arange(0.499, 0, -0.001)

for n_val in n_val:
    p_vals = list(range(2 + excep_len, n_val +1))
    q_vals = list(range(2, n_val + 1 - excep_len))
    for i in range(len(p_vals)):
        for j in range(len(k_o_vals)):
            sim = Ranking_exp(n=n_val, k_o=k_o_vals[j], k_s=1, k_d=0, p=p_vals[i], q=q_vals[i], c_var=0)
            ranks = sim.calc_rank_analytic()  # Get rankings for current n
            if ranks[q_vals[i]-2] - ranks[p_vals[i]-1] < 0:
                if p_vals[i] == n_val: #end of list
                    print('exceptional')
                print(ranks[q_vals[i]-2] - ranks[p_vals[i]-1])
                x_vals = np.arange(1, n_val + 1)
                plt.plot(x_vals, ranks, label=f'p = {p_vals[i]}, q = {q_vals[i]}')
                plt.plot()

                # Add labels and legend
                plt.title(f'Ranking systems by exception pos, k_o = {k_o_vals[j]}, n = {n_val}')
                plt.xlabel('Rankings by exception pos')
                plt.ylabel('Ranking Values')

                plt.legend()
                plt.grid(True)
                plt.xticks(np.arange(1, n_val + 1))  # X-axis ticks every 1

                # Show the plot
                plt.show()

                break

            if j == len(k_o_vals) - 1:
                print(f'No exception for q-1 > p, with q = {q_vals[i]}, p = {p_vals[i]}, n = {n_val}')
                break
```
